# Circular Arc Airfoil Data

Many indoor models use circular arc airfoils. In order to conduct aerodynamic studies of these models, we need standard coefficient data. Unfortunately, research into flight in the range of our indoor models is pretty scarce. However, recent interest in micro UAV vehicles has led to some recent research that is solving this deficiency in data.

I found data for 3% and 5% circular arc airfoils, but that data is not totally applicable for our flights. It will serve as a basis for this study, at least until I find better data.

Using the procedure discussed earlier for digitizing data, we will look at the coefficient data I found.

## Coefficient Data for 3% Arc

```python
import mmtime_helper # to access project code and data
```

```python
from mmtime.utils import get_points, show_curve, fit_curve
```

```python
import os
from pathlib import Path
projdir = Path('/Users/rblack/_dev/nffs-symposium/live-2022')
datadir = projdir / 'mmtime' / 'data' / 'airfoils' / 'arc3'
```

### Lift Coefficient $C_l$

```python
clfile = datadir / 'CL.csv'
cl_x, cl_y = get_points(clfile)
show_curve(cl_x, cl_y, "Lift Coefficient", "alpha","Cl")
```

### Drag Coefficient $C_d$

```python
cdfile = datadir / 'CD.csv'
cd_x, cd_y = get_points(cdfile)
show_curve(cd_x, cd_y, "Drag Coefficient", "alpha","Cd")
```

### Moment Coefficient $C_m$

```python
cmfile = datadir / 'CM.csv'
cm_x, cm_y = get_points(cmfile)
show_curve(cm_x, cm_y, "Moment Coefficient", "alpha","Cm")
```

Let's build the lift-drag polar plot

```python
ClFunc = fit_curve(cl_x, cl_y)
CdFunc = fit_curve(cd_x, cd_y)
```

```python
alpha = np.linspace(cl_x[0], cl_x[-1], 50)
p_x = []
p_y = []
for a in alpha:
    p_x.append(CdFunc(a))
    p_y.append(ClFunc(a))
show_curve(p_x, p_y, "Polar Plot", "Cd", "Cl")
    
```

Cl/Cd PLot

```python
Cl = ClFunc(alpha)
Cd = CdFunc(alpha)
lod = Cl / Cd
show_curve(alpha, lod, "Lift/Drag", "alpha", "L/D")
```
