# XFoil Testing

In this note, we will recreate the research of Michael Reid who explored thin airfoils with a reflex trailing edge using XFoil. The first part of this effort will involve recreating MATLAB code in Python.

```python
import mmtime_helper
```

```python
from mmtime.ReflexAirfoil import ReflexAirfoil
```

```python
ra = ReflexAirfoil(6,  25, 1, 85, 1)
```

```python
from matplotlib import pyplot as plt
xtl,ytl,xbl,ybl = ra.getLE()
plt.figure(figsize=(5,10))
plt.xlim(0, 0.01)
plt.ylim(-0.01, 0.01)
plt.plot(xtl,ytl)
plt.plot(xbl,ybl)
plt.show()
```

```python
xtt,ytt,xbt,ybt = ra.getTE()
plt.figure(figsize=(5,5))
plt.xlim(0.94, 1.01)
plt.ylim(-0.015, 0.005)
plt.plot(xtt,ytt)
plt.plot(xbt,ybt)
plt.show()
```

```python
import math
xt,yt,xb,yb = ra.getCamberPoints()
plt.figure(figsize=(15,1.5))
plt.xlim(-0.01, 1.01)
plt.ylim(-0.02, 0.08)
plt.plot(xtl,ytl)
plt.plot(xbl,ybl)
plt.plot(xt,yt)
plt.plot(xb,yb)
plt.plot(xtt,ytt)
plt.plot(xbt,ybt)
plt.show()
```

```python
nx = 20
t = 0.05# thickness
c = 1.0 - t/2 # chord ( subtract round LE)
r = c**2/(8.0*t) + t/2
theta = 2*math.atan(2*t/c)
r
```

```python
theta * 180/math.pi
```

```python
r * math.sin(theta)
```

```python
r - r * math.cos(theta)
```

```python
dx = (1.0-c/2)/nx
xp = []
yp = []
xc = 0.5
yc = c - r
x = 0
for i in range(nx + 1):
    y = yc + math.sqrt(r**2 - (x - xc)**2)
    xp.append(x)
    yp.append(y)
    x += dx
print(r)
```

```python
#plt.figure(figsize=(15,1.5))
#plt.xlim(-0.01, 1.01)
#plt.ylim(-0.02, 0.08)
plt.plot(xp,yp)
plt.show()
```
