Now, we set up for generating plots:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Grab some data to plot, the lift coefficient for now:

```python
from mmflight.arc3 import CL, CD
xl = CL.x
yl = CL.y
```

BGenerating a basic plot is easy. We use the **plt** tool and tell it to plot some **x-y** points. It will figure out ow to do this. The **show()** action makes it appear. Since we are using *Jupyter*, it will appear as a result of processing this block of code:

To demonstrate how easy it is to generate a basic plot, here is all the code we need.

```python
plt.plot(xl,yl)
plt.show()
```

Now you can see how bad my digitizing job was! we will deal with that later. For now, we need to jazz this us a bit to make it more useful. Let's add some labels and a legend:

```python
plt.plot(xl,yl)
plt.title("Lift Coefficient")
plt.xlabel("angle of attack")
plt.ylabel("Cl")
plt.legend(["3% circular arc"],loc='right')
plt.show()
```

Now, lets show more than one line. We need drag coefficient data to show:

```python
xd = CD.x
yd = CD.y
```

```python

plt.plot(xl,yl)
plt.plot(xd,yd)
plt.title("Lift  and Drag Coefficients")
plt.xlabel("angle of attack")
plt.ylabel("Cl,Cd")
plt.legend(["lift","drag"],loc='lower right')
plt.show()
```
